```python
# â€¢ Selezioniamo i dati relativi a diverse fasce di etÃ : <20, 20-30, 30-40, 40-50, >50; 
filtro_minori_20 = diabetes['Age (years)'] < 20                                         #NELLA TABELLA NON SONO PRESENTI MINORI DI 20
filtro_20_e_30 = (diabetes['Age (years)'] >= 20) & (diabetes['Age (years)'] < 30)
filtro_30_e_40 = (diabetes['Age (years)'] >= 30) & (diabetes['Age (years)'] < 40)
filtro_40_e_50 = (diabetes['Age (years)'] >= 40) & (diabetes['Age (years)'] <= 50)      #HO PROVATO A USARE IL METODO BETWEEN MA NON CI SONO RIUSCITO
filtro_maggiori_50 = diabetes['Age (years)'] >= 50

print(diabetes[filtro_minori_20].shape[0])
print(diabetes[filtro_20_e_30].shape[0])
print(diabetes[filtro_30_e_40].shape[0])
print(diabetes[filtro_40_e_50].shape[0])
print(diabetes[filtro_maggiori_50].shape[0])
```

```python
# â€¢ Qual Ã¨ la media della pressione sanguigna diastolica per le diverse fasce di etÃ ?
filtro_minori_20 = diabetes['Age (years)'] < 20

maschera_minori_20 = diabetes.loc[filtro_minori_20]

#media_minori_20 = round(maschera_minori_20["Diastolic blood pressure (mm Hg)"].mean())
#---------------------------------------------------------------------------------------------
filtro_20_e_30 = (diabetes['Age (years)'] >= 20) & (diabetes['Age (years)'] < 30)       #FILTRO FASCIA DI ETA DAL PRECEDENTE ESERCIZIO

maschera_20_30 = diabetes.loc[filtro_20_e_30]                                           #MASCHERA PER TABELLA

media_20_e_30 = round(maschera_20_30["Diastolic blood pressure (mm Hg)"].mean())     #MEDIA DELLA PRESSIONE SANGUIGNA BASATA SULLA MASCHERA
#---------------------------------------------------------------------------------------------
filtro_30_e_40 = (diabetes['Age (years)'] >= 30) & (diabetes['Age (years)'] < 40)

maschera_30_e_40 = diabetes.loc[filtro_30_e_40]

media_30_e_40 = round(maschera_30_e_40["Diastolic blood pressure (mm Hg)"].mean())
#---------------------------------------------------------------------------------------------
filtro_40_e_50 = (diabetes['Age (years)'] >= 40) & (diabetes['Age (years)'] <= 50) 

maschera_40_e_50 = diabetes.loc[filtro_40_e_50]

media_40_e_50 = round(maschera_40_e_50["Diastolic blood pressure (mm Hg)"].mean())
#---------------------------------------------------------------------------------------------
filtro_maggiori_50 = diabetes['Age (years)'] > 50

maschera_maggiori_50 = diabetes.loc[filtro_maggiori_50]

media_maggiori_50 = round(maschera_maggiori_50["Diastolic blood pressure (mm Hg)"].mean())
#---------------------------------------------------------------------------------------------
print("Nella lista non sono presenti minori di 20")
print("La media per la fascia di etÃ  20-30 Ã¨ di:" , media_20_e_30)
print("La media per la fascia di etÃ  30-40 Ã¨ di:" , media_30_e_40)
print("La media per la fascia di etÃ  40-50 Ã¨ di:" , media_40_e_50)
print("La media per la fascia di etÃ  superiori ai 50 Ã¨ di:" , media_maggiori_50)
```

```python
# â€¢ Qual Ã¨ la media della pressione per ogni singolo anno di etÃ ?
media_pressione_x_eta = diabetes.groupby(['Age (years)'])["Diastolic blood pressure (mm Hg)"].mean()
print(media_pressione_x_eta)
```
