**Esercizio 2/3 :

• Quanto costano in media i prodotti (colonna DealerPrice)?

```python
# • In media quanto pesano i prodotti? 
print('La media del peso dei prodotti è:' , round(df['Weight'].mean()))
```

```python
# • Quanto pesa il più leggero? 
print('Il prodotto più leggero è:' ,df['Weight'].min())
```

```python
# • Quanto pesa il più pesante? 
print('Il prodotto più pesante è:' ,df['Weight'].max())
```

```python
# • Quanti prodotti pesano più di 100 Kg? 
filtro = df['Weight'] > 100 
df.loc[filtro].shape[0]
```

```python
# • Quanto costano in media i prodotti (colonna DealerPrice)? 
print('La media del costo dei prodotti è:' , round(df['DealerPrice'].mean()))
```

```python
# • Se prendiamo un quarto di tutti i prodotti, in modo che siano i più costosi, quale sarà il loro range di prezzo?
quarto = round(len(df['DealerPrice']) / 4)    #QUANTITA TOT DI RIGHE DIVISO 4 ARROTONDATO

riordino = df.sort_values(by = 'DealerPrice' , ascending = False)       #RIORDINO DELLA LISTA IN MODO DISCENDENTE SICCOME IN ALTO CI SARANNO I VALORI NaN

maschera = riordino['DealerPrice'].head(quarto)                 #HO PRESO L'ULTIMO QUARTO

print("Il range va da:" , maschera.min() , "fino a:" , maschera.max())  
```

Esercizio 3/3 : 
• Qual è il prezzo medio per i prodotti di colore blu? 
• Qual è il prezzo medio per i prodotti di colore rosso o nero? 
• Qual è il prezzo massimo per i prodotti di taglia 42 e peso oltre i 10 Kg? 
• Qual è il nome inglese e il costo di produzione (StandardCost) di tutti i prodotti di taglia 42, peso oltre i 10 Kg e colore argento? 
• Esaminiamo la differenza tra lo StandardCost e il DealerPrice degli ultimi 20 elementi del dataset: c'è un pattern? C'è qualche elemento che non lo segue?

```python
# • Qual è il prezzo medio per i prodotti di colore blu? 
filtro_blu = df['Color'] == 'Blue'
filtro_prezzi = df['DealerPrice']

maschera = df.loc[(filtro_blu) & (filtro_prezzi)]                      #HO CREATO LA MASCHERA PER UNA TABELLA DOVE CI SONO SOLTANTO COLORI BLU

somma = sum(maschera['DealerPrice'])                                    #HO FATTO LA MEDIA DI ESSI
conteggio = len(maschera['DealerPrice']) 
media = somma / conteggio

print(media)
```

```python
# • Qual è il prezzo medio per i prodotti di colore rosso o nero? 
filtro_rosso = df['Color'] == 'Red'
filtro_nero = df['Color'] == 'Black'

maschera = df.loc[(filtro_rosso) | (filtro_nero)]                                   #HO CREATO LA MASCHERA PER UNA TABELLA DOVE CI SONO SOLTANTO COLORI BLU
#maschera = maschera.fillna(False)                         #NELLA COLONNA PREZZI ERANO PRESENTI DEI NAN CHE HO SOSTITUITO CON ZERI

#somma = sum(maschera['DealerPrice'])
#conteggio = len(maschera['DealerPrice'])
#media = somma / conteggio

print(maschera['DealerPrice'].mean())
```

```python
# • Qual è il prezzo massimo per i prodotti di taglia 42 e peso oltre i 10 Kg? 
filtro_taglia = df['Size'] == '42'
filtro_peso = df['Weight'] > 10

maschera = df[(filtro_taglia) & (filtro_peso)]
maschera['DealerPrice'].max()
```

```python
# • Qual è il nome inglese e il costo di produzione (StandardCost) di tutti i prodotti di taglia 42, peso oltre i 10 Kg e colore argento? 
filtro_taglia = df['Size'] == '42'
filtro_peso = df['Weight'] > 10
filtro_colore = df['Color'] == 'Silver'

maschera = df[(filtro_taglia) & (filtro_peso) & (filtro_colore)]
maschera.loc[: , ['EnglishProductName' , 'StandardCost' , 'Color' , 'Size' , 'Weight']]
```

```python
# • Esaminiamo la differenza tra lo StandardCost e il DealerPrice degli ultimi 20 elementi del dataset: 
# c'è un pattern? C'è qualche elemento che non lo segue?
df.loc[: , ['StandardCost' , 'DealerPrice']].tail(20)
```
