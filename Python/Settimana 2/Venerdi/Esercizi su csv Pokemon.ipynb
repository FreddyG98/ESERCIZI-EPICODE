**Esercizio 2/2 
• Quali sono i Pokémon leggendari? 
• E quali sono i leggendari di tipo 1 Grass? 
• E leggendari di tipo 1 Ice o Fire? 
• Ordiniamo il dataset per la colonna Name; 
• Trasformiamo Name nell'indice; 
• Quali sono i Pokémon della prima generazione con attacco > 50 e HP < 60?**

```python
# • Quali sono i Pokémon leggendari? 
filtro_leggendari = pokemon['Legendary'] == True
pokemon[filtro_leggendari]
```

```python
# • E quali sono i leggendari di tipo 1 Grass? 
filtro_leggendari = pokemon['Legendary'] == True
filtro_grass = pokemon['Type 1'] == 'Grass'

pokemon[(filtro_leggendari) & (filtro_grass)]
```

```python
# • E leggendari di tipo 1 Ice o Fire? 
filtro_leggendari = pokemon['Legendary'] == True
filtro_ice = pokemon['Type 1'] == 'Ice'
filtro_fire = pokemon['Type 1'] == 'Fire'

pokemon[(filtro_leggendari) & ((filtro_ice) | (filtro_fire))]
```

```python
# • Ordiniamo il dataset per la colonna Name; 
pokemon.sort_values(by = 'Name')
```

```python
# • Trasformiamo Name nell'indice; 
pokemon = pokemon.sort_values(by = 'Name').set_index('Name')
pokemon
```

```python
# • Quali sono i Pokémon della prima generazione con attacco > 50 e HP < 60?
filtro_generazione = pokemon['Generation'] == 1
filtro_attacco = pokemon['Attack'] > 50
filtro_hp = pokemon['HP'] < 60

pokemon[(filtro_generazione) & ((filtro_attacco) & (filtro_hp))]
```
