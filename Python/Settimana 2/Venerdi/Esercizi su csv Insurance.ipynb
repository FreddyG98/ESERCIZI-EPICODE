```python
# • Visualizziamone le dimensioni, un'anteprima, e osserviamo i nomi di colonna; 
print("Dimensioni:" , insurance.shape , "\n")
insurance.info()
```

```python
insurance.sample(10)
```

```python
# • Quali sono le medie di charges rispetto a region? Ci sono differenze significative? 
media_charges_x_regione = insurance.groupby(['region'])['charges'].mean()
print(media_charges_x_regione)
```

```python
# • E rispetto a smoker? E a sex? 
m_charges_smoker_e_sex = insurance.groupby(['region' , 'smoker' , 'sex'])['charges'].mean()
print(m_charges_smoker_e_sex)
```

```python
# • Quali sono i descrittori statistici di bmi? 
insurance['bmi'].describe()
```

```python
# • Quali sono minimo, media e massimo di charges rispetto ai diversi quartili dei valori di bmi?
insurance['bmi quartili'] = pd.qcut(insurance['bmi'] , q = 4 , labels = ['Q1' , 'Q2' , 'Q3' , 'Q4'])

insurance[['bmi' , 'bmi quartili']]

insurance.groupby('bmi quartili')['charges'].agg(['min' , 'mean' , 'max'])
```
